## 2. Run Translation

```python
# Change target language: kannada, tamil, telugu, hindi, bengali, gujarati, malayalam, odia, punjabi, etc.
TARGET = "kannada"

# Optional: limit chapters for quick test (remove or set to 0 for full)
LIMIT = 0  # e.g. 5 for testing
```

```python
cmd = f"python translate_pipeline.py --target {TARGET} --backend indic_trans2"
if LIMIT:
    cmd += f" --limit {LIMIT} --no-phrases --no-dictionary"
!{cmd}
```

## 3. Download Output Files

```python
from google.colab import files
import os

for f in [f"data_{TARGET}.json", f"phrases_{TARGET}.json", f"dictionary_{TARGET}.json"]:
    if os.path.exists(f):
        files.download(f)
        print(f"Downloaded {f}")
    else:
        print(f"Not found: {f}")
```

## Multiple Languages (Optional)

```python
# Download all output files
from google.colab import files
import os

for lang in LANGUAGES:
    for suffix in ["data", "phrases", "dictionary"]:
        f = f"{suffix}_{lang}.json"
        if os.path.exists(f):
            files.download(f)
            print(f"Downloaded {f}")
```
